# Image Segmentation

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

def display_image(image, title="Image"):
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

def process_document(img_file, layout_type="three_columns"):
    # Read the image
    image = cv2.imread(img_file)
    base_image = image.copy()

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply Gaussian blur
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # Thresholding
    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    
    # Erode to help separate columns before dilation
    erode_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
    eroded = cv2.erode(thresh, erode_kernel, iterations=1)

    # Dilation for finding columns
    if layout_type == "multi_columns":
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 20))  # Adjusted kernel size for better column detection
    else:  # Single column layout
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (50, 5))
    
    dilate = cv2.dilate(eroded, kernel, iterations=1)  # Adjusted iterations for balanced dilation
    
    # Find contours
    cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    cnts = sorted(cnts, key=lambda x: cv2.boundingRect(x)[0])  # Sorting by x coordinate for column ordering
    
    # Draw bounding boxes and extract regions of interest (ROI)
    for i, c in enumerate(cnts):
        x, y, w, h = cv2.boundingRect(c)
        if layout_type == "multi_columns":
            if h > 110 and w > 50:
                # roi = base_image[y:y+h, x:x+w]
                # ocr_result = pytesseract.image_to_string(roi)
                # print(ocr_result)
                # cv2.imwrite("temp/boundrybox"+str(i)+".JPG",roi)
                cv2.rectangle(image, (x-5, y), (x+w+5, y+h), (36, 255, 12), 2)
        else:  # Single column layout
            if h > 50 and w > 200:
                # roi = base_image[y:y+h, x:x+w]
                cv2.rectangle(image, (x, y), (x+w, y+h), (36, 255, 12), 2)
    
    # Display the processed image with bounding boxes
    display_image(image, title="Detected Columns/Regions")

    # Perform OCR on the original image
    ocr_result_original = pytesseract.image_to_string(base_image)
    
    # Print the OCR result
    print("OCR Result:\n", ocr_result_original)

    return ocr_result_original

# Example usage
img_file = "../data/two_columns.jpeg"  # Replace with the path to your image
ocr_result = process_document(img_file, layout_type="multi_columns")
# For single column documents use:
# ocr_result = process_document(img_file, layout_type="single_column")
```
